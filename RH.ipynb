# PrÃ©-processamento e bases de treinamento/teste

```python
#tratando atributos categoricos

x_cat = employee_df[['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']]
x_cat
```

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()
x_cat = onehotencoder.fit_transform(x_cat).toarray()
```

```python
x_cat = pd.DataFrame(x_cat)
x_cat
```

```python
employee_df.info()
```

```python
# Selecionar apenas as colunas que tÃªm tipo de dados numÃ©rico, exceto 'Attrition'
x_numerical = employee_df.select_dtypes(include=['number']).drop(columns=['Attrition'])
x_numerical
```

```python
x_all = pd.concat([x_cat, x_numerical], axis = 1)
x_all
```

```python
x_all.info()
```

```python
# Normalizando todas as colunas do DataFrame
from sklearn.preprocessing import MinMaxScaler

x_all.columns = x_all.columns.astype(str)
scaler = MinMaxScaler()
x_normalized = scaler.fit_transform(x_all)
```

```python
y= employee_df['Attrition']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_normalized,y, test_size=0.25)
```

```python
x_train.shape
```

# RegressÃ£o LogÃ­stica

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic.fit(x_train, y_train)
```

```python
y_pred = logistic.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
```

```python
# Calcular a precision
precision = precision_score(y_test, y_pred)
print("Precision:", precision)

# Calcular o recall
recall = recall_score(y_test, y_pred)
print("Recall:", recall)

# Calcular o F1-score
f1 = f1_score(y_test, y_pred, average='macro')
print("F1-score:", f1)
```

```python
print(classification_report(y_test, y_pred))
```

```python
report = classification_report(y_test, y_pred, output_dict=True)

recall_0 = report['0']['recall']
recall_1 = report['1']['recall']

description = f"O modelo classifica corretamente {recall_0*100:.0f}% das pessoas que ficarÃ£o na empresa, porem classifica corretamente somente {recall_1*100:.0f}% das pessoas que sairÃ£o da empresa."

print(description)
```
